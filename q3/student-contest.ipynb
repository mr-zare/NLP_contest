```python
# Reading/Loading the dataset files
# TODO
df=pd.read_csv("./data/train.csv")
df_test=pd.read_csv("./data/test.csv")
x=df.drop(columns=['GPA','StudentID'])
y=df['GPA']
df
```

        در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی دلخواهتان، استفاده کنید.

    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.

```python
# Preprocessing step
# TODO
from sklearn.impute import SimpleImputer
```

```python
cols=['ParentalEducation','ParentalSupport']
imput=SimpleImputer(strategy='most_frequent')
x[cols]=imput.fit_transform(x[cols])
x
```

```python
columns=['Age', 'Gender', 'Ethnicity', 'ParentalEducation',
       'StudyTimeWeekly', 'Absences', 'Tutoring', 'ParentalSupport',
       'Extracurricular', 'Sports', 'Music', 'Volunteering']
print(x[columns].isna().sum())
```

```python
from sklearn.preprocessing import OrdinalEncoder
category_columns = ['Ethnicity', 'ParentalEducation', 'ParentalSupport']
encode=OrdinalEncoder()
x[category_columns]=encode.fit_transform(x[category_columns])
x['all_extrac']=x[['Extracurricular', 'Sports', 'Music', 'Volunteering']].sum(axis=1)
x
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
آموزش مدل
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.
</font>
</p>

```python
from sklearn.model_selection import train_test_split

x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
from sklearn.metrics import r2_score, mean_squared_error
```

```python
param_grid = {
    'n_estimators': [100, 200, 300,800],
    'learning_rate': [0.01, 0.1,0.001,0.05],
    'max_depth': [2,3, 5, 7,9,13]
}

grid_search = GridSearchCV(estimator=xgb.XGBRegressor(), param_grid=param_grid, scoring='r2', cv=5)

grid_search.fit(x_train, y_train)

print(f'Best Hyperparameters: {grid_search.best_params_}')

best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(x_val)

r2_best = r2_score(y_val, y_pred_best)
mse_best = mean_squared_error(y_val, y_pred_best)
print(f'Best Model R2 Score: {r2_best}')
print(f'Best Model MSE: {mse_best}')
```

```python
df_test=df_test.drop(columns=['StudentID'])
cols=['ParentalEducation','ParentalSupport']
imput=SimpleImputer(strategy='most_frequent')
df_test[cols]=imput.fit_transform(df_test[cols])
category_columns = ['Ethnicity', 'ParentalEducation', 'ParentalSupport']
encode=OrdinalEncoder()
df_test[category_columns]=encode.fit_transform(df_test[category_columns])
df_test['all_extrac']=x[['Extracurricular', 'Sports', 'Music', 'Volunteering']].sum(axis=1)
```

```python
import zipfile
import os
if not os.path.exists(os.path.join(os.getcwd(), 'Student_GPA.ipynb')):
    %notebook -e Student_GPA.ipynb

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)

file_names = ['Student_GPA.ipynb', 'submission.csv']
compress(file_names)
```
